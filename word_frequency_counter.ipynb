```python
import re
from matplotlib import pyplot as plt
from itertools import islice

frequency = {}
document_text = open('research_driven_critique_essay.txt', 'r', encoding='utf8')
text_string = document_text.read().lower()
match_pattern = re.findall(r'\b[a-z]{3,15}\b', text_string)

for word in match_pattern:
    count = frequency.get(word,0)
    frequency[word] = count + 1
most_frequent = dict(sorted(frequency.items(), key=lambda elem: elem[1], reverse=True))
most_frequent_count = most_frequent.keys()

for words in most_frequent_count:
    print(words, '____: ', most_frequent[words])


# generating graph and changing the size to 20x20.
plt.figure(figsize=(20, 20))
plt.title("Words density")
plt.xlabel("words")
plt.ylabel("frequency")
# plots a bar graph with X as the KEYS of the dictionary and Y as the VALUES of the dictionary
plt.bar(list(most_frequent.keys())[:25], list(most_frequent.values())[:25], width = 0.4)
plt.show()
```
